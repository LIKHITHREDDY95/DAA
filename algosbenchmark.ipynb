```python
import time
import random
import matplotlib.pyplot as plt


def bubble_sort(unsorted_array):
    arr_length = len(unsorted_array)
    outer_index = 0

    while outer_index < arr_length:
        no_of_swaps_occurred = False
        inner_index = 0

        while True:
            if inner_index >= arr_length - outer_index - 1:
                break

            current_element = unsorted_array[inner_index]
            next_element = unsorted_array[inner_index + 1]

            if current_element > next_element:
                unsorted_array[inner_index], unsorted_array[inner_index + 1] = next_element, current_element
                no_of_swaps_occurred = True

            inner_index += 1

        if not no_of_swaps_occurred:
            return unsorted_array

        outer_index += 1

    return unsorted_array


if __name__ == "__main__":
    unsorted_sequence = [64, 34, 25, 12, 22, 11, 90]
    print("unsorted array:", unsorted_sequence)
    sorted_sequence = bubble_sort(unsorted_sequence)
    print("sorted array:", sorted_sequence)


def bubble_sort_time(size):
    arr = [random.randint(1, 1000) for _ in range(size)]
    start_time = time.time()
    bubble_sort(arr)
    end_time = time.time()
    return end_time - start_time


# Array sizes to test
sizes = range(5,10000, 800)
times = []

# Measure time for each array size
for size in sizes:
    time_taken = bubble_sort_time(size)
    times.append(time_taken)
    print(f"Array size: {size}, Time taken: {time_taken:.6f} seconds")

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(sizes, times, marker='+',color='r')
plt.title('Bubble Sort: Array Size vs Time Taken')
plt.xlabel('Array Size')
plt.ylabel('Time Taken (seconds)')
plt.grid(True)
plt.show()



def selection_sort(sequence):
    sequence_length = len(sequence)
    index = 0

    while True:
        if index >= sequence_length - 1:
            break

        minimum_value = sequence[index]
        minimum_index = index
        inner_index = index + 1

        while inner_index < sequence_length:
            if sequence[inner_index] < minimum_value:
                minimum_value = sequence[inner_index]
                minimum_index = inner_index
            inner_index += 1

        if minimum_index != index:
            sequence[index], sequence[minimum_index] = sequence[minimum_index], sequence[index]

        index += 1

    return sequence


if __name__ == "__main__":
    unsorted_list = [683, 1324, 34, -3, -2, -0.098, 10, 0.8828, 0.000000323]
    print("Unsorted list:", unsorted_list)
    sorted_list = selection_sort(unsorted_list)
    print("Sorted list:", sorted_list)

# List of input sizes
input_sizes = range(5, 10000 ,700)

# List to store times taken
times_taken = []

# Loop over each input size
for size in input_sizes:
    # Generate a random list of the current size
    arr = [random.randint(0, 10000) for _ in range(size)]

    # Time the selection sort
    start_time = time.time()
    selection_sort(arr)
    end_time = time.time()

    # Record the time taken
    times_taken.append(end_time - start_time)

# Plotting the results
plt.plot(input_sizes, times_taken, marker='o' , color='y')
plt.title('Number of Inputs vs Time Taken for Selection Sort')
plt.xlabel('Number of Inputs')
plt.ylabel('Time Taken (seconds)')
plt.grid(True)
plt.show()



def insertion_sort(data_sequence):
    seq_length = len(data_sequence)
    sort_index = 1

    while sort_index < seq_length:
        value = data_sequence[sort_index]
        shift_index = sort_index - 1

        while True:
            if shift_index < 0 or data_sequence[shift_index] <= value:
                break

            data_sequence[shift_index + 1] = data_sequence[shift_index]
            shift_index -= 1

        data_sequence[shift_index + 1] = value

        sort_index += 1

    return data_sequence


if __name__ == "__main__":
    unsorted_sequence = [13, 5, -98 , 0 , 0.7389 , -0.367438 , -86743 , -98 , 235 , 7548 , 101893]
    print("Unsorted sequence:", unsorted_sequence)
    sorted_sequence = insertion_sort(unsorted_sequence)
    print("Sorted sequence:", sorted_sequence)
def measure_time(func, arr):
    start_time = time.time()
    func(arr)
    end_time = time.time()
    return end_time - start_time

sizes = range(5,9000 , 1000)
times = []

for size in sizes:
    arr = [random.randint(1, 1000) for _ in range(size)]
    execution_time = measure_time(insertion_sort, arr.copy())
    times.append(execution_time)

plt.figure(figsize=(10, 6))
plt.plot(sizes, times, marker='+' , color='g')
plt.title('Insertion Sort: Array Size vs Time')
plt.xlabel('Array Size')
plt.ylabel('Time (seconds)')
plt.grid(True)
plt.show()
```
